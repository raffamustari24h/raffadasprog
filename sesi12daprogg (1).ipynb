```python
# File: program_interaktif.py
"""
Program interaktif dengan fitur Aritmatika, Konversi, dan Ubah Bilangan
dalam satu berkas, menggunakan menu pilihan.
"""

def penjumlahan(a, b):
    return a + b

def perpangkatan(a, b):
    return a ** b

def perkalian(a, b):
    return a * b

def cm_ke_m(cm):
    return cm / 100.0

def m_ke_cm(m):
    return m * 100.0

def desimal_ke_biner(n):
    return bin(n)[2:]

def desimal_ke_oktal(n):
    return oct(n)[2:]

def desimal_ke_hex(n):
    return hex(n)[2:].upper()

def menu_aritmatika():
    print("\n-- Submenu Aritmatika --")
    print("1. Penjumlahan")
    print("2. Perpangkatan")
    print("3. Perkalian")
    pilihan = input("Pilih operasi (1-3): ")
    try:
        op = int(pilihan)
        a = float(input("Masukkan bilangan pertama: "))
        b = float(input("Masukkan bilangan kedua: "))
        if op == 1:
            print(f"Hasil penjumlahan: {penjumlahan(a, b)}")
        elif op == 2:
            print(f"Hasil perpangkatan: {perpangkatan(a, b)}")
        elif op == 3:
            print(f"Hasil perkalian: {perkalian(a, b)}")
        else:
            print("Pilihan tidak valid.")
    except ValueError:
        print("Input tidak valid.")

def menu_konversi():
    print("\n-- Submenu Konversi --")
    print("1. CM ke M")
    print("2. M ke CM")
    pilihan = input("Pilih konversi (1-2): ")
    try:
        op = int(pilihan)
        nilai = float(input("Masukkan nilai desimal: "))
        if op == 1:
            print(f"{nilai} cm = {cm_ke_m(nilai)} m")
        elif op == 2:
            print(f"{nilai} m = {m_ke_cm(nilai)} cm")
        else:
            print("Pilihan tidak valid.")
    except ValueError:
        print("Input tidak valid.")

def menu_ubah_bilangan():
    print("\n-- Submenu Ubah Bilangan --")
    print("1. Desimal ke Biner")
    print("2. Desimal ke Oktal")
    print("3. Desimal ke Hexadesimal")
    pilihan = input("Pilih konversi (1-3): ")
    try:
        op = int(pilihan)
        nilai = int(input("Masukkan nilai desimal (integer): "))
        if op == 1:
            print(f"Desimal {nilai} ke Biner = {desimal_ke_biner(nilai)}")
        elif op == 2:
            print(f"Desimal {nilai} ke Oktal = {desimal_ke_oktal(nilai)}")
        elif op == 3:
            print(f"Desimal {nilai} ke Hexadesimal = {desimal_ke_hex(nilai)}")
        else:
            print("Pilihan tidak valid.")
    except ValueError:
        print("Input tidak valid.")

def main():
    while True:
        print("\n=== Menu Utama ===")
        print("1. Aritmatika")
        print("2. Konversi")
        print("3. Ubah Bilangan")
        print("4. Keluar")
        pilih = input("Pilih menu (1-4): ")
        if pilih == '1':
            menu_aritmatika()
        elif pilih == '2':
            menu_konversi()
        elif pilih == '3':
            menu_ubah_bilangan()
        elif pilih == '4':
            print("Terima kasih!")
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```
