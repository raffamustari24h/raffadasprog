```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  

np.random.seed(42)

bulan = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']  # Fixed 'no' to 'nov' and 'des' to 'dec'

minggu1 = np.random.randint(20000, 500001, 12)
minggu2 = np.random.randint(20000, 5000001, 12)
minggu3 = np.random.randint(20000, 500001, 12)
minggu4 = np.random.randint(20000, 500001, 12)

data = {
    'bulan': bulan,  
    'minggu1': minggu1,
    'minggu2': minggu2,
    'minggu3': minggu3,
    'minggu4': minggu4
}

df = pd.DataFrame(data)  
df['fatal_pengeluaran'] =df[['minggu1','minggu2','minggu3','minggu4']].sum(axis=1)
df 
```

```python
plt.figure(figsize=(12,6))
plt.bar(df['bulan'],df['fatal_pengeluaran'],color='yellow')
plt.title("penguaran")
plt.xticks(rotation=45)
plt.ylabel("rupiah")
plt.xlabel("bulan")
plt.grid(axis='y')
plt.tight_layout()


plt.show()
```

```python
plt.figure(figsize=(12,8))

for minggu in ['minggu1','minggu2','minggu3','minggu4']:
    plt.plot(df['bulan'],df[minggu],marker='o',label=minggu)

plt.title('penguaran minggu')
plt.xlabel('bulan')
plt.ylabel('rupiah')
plt.legend()
plt.grid(True)
plt.xticks(rotation = -45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.title("proporsi")
plt.pie(df['fatal_pengeluaran'], labels=df['bulan'], startangle=90, autopct="%1.1f%%")
plt.axis('equal')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
df['bulan_angka'] = np.arange(1,13)

x = df[['bulan_angka']]
y = df['fatal_pengeluaran']

model = LinearRegression()
model.fit(x,y)

prediksi = model.predict([[13]])[0]
print(f"Pengeluaran di bulan 13: {prediksi:,.0f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("https://datahub.io/core/corruption-perceptions-index/_r/-/data/cpi.csv")
df_filtered = df[['Jurisdiction', '2015']].dropna()  
df_filtered = df_filtered.sort_values(by='2015', ascending=False)

plt.figure(figsize=(12,8))

plt.bar(df_filtered['Jurisdiction'][:10], df_filtered['2015'][:10], color='blue')
plt.xlabel("Negara")
plt.ylabel("Indeks Persepsi Korupsi")
plt.title("10 Negara dengan Indeks Persepsi Korupsi Tertinggi (2015)")
plt.xticks(rotation=45)
plt.show()


```
